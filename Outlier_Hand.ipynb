```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
sns.boxplot(df)
plt.show()
```

```python
list1=list(df.columns)           # List of features 
for i in list1:                  # To iterate for each feature
    q1 = df[i].quantile(0.25)
    q2 = df[i].quantile(0.5)
    q3 = df[i].quantile(0.75)
    
    iqr = q3 - q1
    upper_tail = q3 + 1.5* iqr
    lower_tail = q1 - 1.5* iqr
    
    df[i]  = np.where((df[i] < lower_tail) | (df[i] > upper_tail) ,df[i].median(), df[i])
    
```

```python
sns.boxplot(df)
plt.show()

```
